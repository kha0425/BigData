### 작업형 - 제 1유형

01. 다음은 California Housing 데이터 세트이다.
 데이터 중 결측치가 있는 경우 해당 데이터의 행을 모두 제거하고, 첫 번째 행부터 순서대로 70%까지의 데이터를 훈련 데이터로 추출한 데이터 세트를 구성한다. 변수 중 'housing_median_age'의 Q1(제1사분위수)값을 정수로 계산하시오.

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
df1 = pd.read_csv('./data/2_1_housing.csv')
df1.head()
```

```python
# 결측치 확인
df1.info()
```

```python
# 결측치가 있는 해당 데이터의 행 제거
df_dropna = df1.dropna()
print(len(df1))
print(len(df_dropna))
```

```python
# 70% 데이터 추출
df_dropna_70 = df_dropna.iloc[:int(len(df_dropna)*0.7)]
print(len(df_dropna_70))
```

```python
# 'housing_median_age' 제 1사분위수 정수 계산
result = int(df_dropna_70['housing_median_age'].quantile(0.25))
print(result)
```

2. 다음은 국가별 연도별 인구 10만 명당 결핵 유병률 데이터 세트이다. 2000년도의 국가별 결핵 유병률 데이터 세트에서 2000년도의 평균값보다 더 큰 유병률 값을 가진 국가의 수를 계산하시오.

```python
# 라이브러리 및 데이터 불러오기
df2 = pd.read_csv('./data/2_2_worlddata.csv')
df2.head()
```

```python
a = df2['year'] == 2000
print(a)
```

```python
# 2000년도의 유병률 평균값
# 2000년도에 해당하는 데이터 추출
df_year_2000 = df2[df2['year']==2000]
print(df_year_2000)
```

```python
# 행과 열 변환
df_year_2000_T = df_year_2000.T
print(df_year_2000_T)
```

```python
# year 행 삭제
df_year_2000_T = df_year_2000_T.drop(['year'])
print(df_year_2000_T)
```

```python
# 컬럼값 지정 후 평균값 계산
df_year_2000_T.columns = ['value']
mean = df_year_2000_T['value'].mean()
print(mean)
```

```python
# 평균값보다 더 큰 유병률 값을 가진 국가의 수 계산
result = len(df_year_2000_T[df_year_2000_T['value']>mean])
print(result)
```

### 작업형 - 제 2유형

01. 다음은 Travel Insurance 데이터 세트이다. 주어진 훈련 데이터 세트를 이용하여 고객별 여행보험 가입 여부 예측 모형을 만들고, 가장 높은 Accuracy 값을 가지는 최종 모델을 도출하시오. 해당 모델을 활용하여 보험 가입 여부 예측값을 계산하고 결괏값은 csv 파일로 제출하시오.

- 결과 제출 양식 : 제출한 예측값의 ROC_AUC 점수 결과를 통해 영역별 배점에 따라 최종 점수가 반영될 예정

- 변수 설명

> ID : 고객 ID 번호

> Age : 고객 나이

> Employment Type : 고객 직업 유형

> GraduateOrNot : 대졸 여부

> Annuallncome : 연 소득

> FamilyMembers : 가족 수

> ChronicDisease : 만성질환 여부

> FrequentFlyer : 비행 단골고객 여부

> EverTravelledAbroad : 해외여행 경험

> TravelInsurance : 여행보험 가입 여부 (0 : 미가입, 1 : 가입)


```python
# 라이브러리 및 데이터 불러오기
train = pd.read_csv('./data/2_1_travel_insurance_train.csv')
test = pd.read_csv('./data/2_2_travel_insurance_test.csv')
```

```python
train.head()
```

```python
test.head()
```

```python
X_train = train.drop(columns = 'TravelInsurance')
y_train = train[['ID','TravelInsurance']]
X_test = test.copy()
```

```python
print(X_train.shape,y_train.shape,X_test.shape)
```

```python
X_train.info()
```

```python
X_test.info()
```

```python
# object 변수 원 - 핫 인코딩
# int 변수 스케일링
int_train = X_train.select_dtypes(exclude = 'object')
object_train = X_train.select_dtypes(include = 'object')
int_test = X_test.select_dtypes(exclude = 'object')
object_test = X_test.select_dtypes(include = 'object')
```

```python
int_train.head()
```

```python
object_train.head()
```

```python
# 수치형변수 스케일링
# 이상치 : RobustScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# ID는 스케일링이 필요없기 때문에 컬럼 리스트 생성
cols = ['Age','AnnualIncome','FamilyMembers','ChronicDiseases']

display(int_train.head())
int_train[cols] = scaler.fit_transform(int_train[cols])
int_test[cols] = scaler.fit_transform(int_test[cols])
int_train.head()
```

```python
# 범주형 데이터 : 원 - 핫 인코딩
display(object_train.head())
object_train = pd.get_dummies(object_train)
object_test = pd.get_dummies(object_test)
object_train.head()
```

```python
# 데이터 결합
X_train = pd.concat([int_train,object_train],axis = 1)
X_test = pd.concat([int_test,object_test],axis = 1)
print(X_train.shape,X_test.shape)
X_train.head()
```

```python
# 검증 데이터 분리
from sklearn.model_selection import train_test_split
X_tr,X_val,y_tr,y_val = train_test_split(X_train,y_train['TravelInsurance'],test_size = 0.1,random_state = 1234)
X_tr.shape,X_val.shape,y_tr.shape,y_val.shape
```

```python
# 모델링 및 평가 라이브러리 임포트
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import roc_auc_score
```

```python
# 로지스틱회귀분석
model_lgm = LogisticRegression()
model_lgm.fit(X_tr,y_tr)
pred_lgm = model_lgm.predict_proba(X_val)
roc_auc_score(y_val,pred_lgm[:,1])
```

```python
# 서포트벡터머신
model_svm = SVC(probability = True)
model_svm.fit(X_tr,y_tr)
pred_svm = model_svm.predict_proba(X_val)
roc_auc_score(y_val,pred_svm[:,1])
```

```python
# 의사결정나무
model_dt = DecisionTreeClassifier()
model_dt.fit(X_tr,y_tr)
pred_dt = model_dt.predict_proba(X_val)
roc_auc_score(y_val,pred_dt[:,1])
```

```python
# 랜덤포레스트
model_rf = RandomForestClassifier()
model_rf.fit(X_tr,y_tr)
pred_rf = model_rf.predict_proba(X_val)
roc_auc_score(y_val,pred_rf[:,1])
```

```python
# LGBM
model_lgb = LGBMClassifier()
model_lgb.fit(X_tr,y_tr)
pred_lgb = model_lgb.predict_proba(X_val)
roc_auc_score(y_val,pred_lgb[:,1])
```

```python
# 최종 모델 선택
model_final = RandomForestClassifier()
model_final.fit(X_train,y_train['TravelInsurance'])
pred_final = model_final.predict_proba(X_test)
pred_final[:5]
```

```python
submission = pd.DataFrame({
    'index': X_test.index,
    'y_pred' : pred_final[:,1]
})
submission.head()
```

```python
# 수험번호.csv
submission.to_csv('./output/submission.csv',index = False)
```
