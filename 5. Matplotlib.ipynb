```python
# 맷플롯립 (Matplotlib)
# 데이터를 차트(Chart)나 플롯(Plot)으로 시각화 하는 패키지
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [2,4,8]
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("Line Graph")
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
plt.plot(x,y)
plt.title("Line Graph", color='blue')
plt.xlabel('X', color = 'red')
plt.ylabel('Y', color = '#00aa00')
plt.show()
```

```python
plt.plot(x,y)
plt.xticks([1,2,3])
plt.yticks([3,6,9,12])
plt.show()
```

```python
plt.plot(x,y,label='lengend') # 범례 지정
plt.legend()
plt.show()
```

```python
plt.plot(x,y,label='score') # 범례 지정
plt.legend()
plt.show()
```

```python
plt.plot(x,y,label='legend')
plt.legend(loc=(0.75,0.9))
plt.show()
```

```python
plt.plot(x,y,color='red',linewidth=3,
         linestyle = ':', marker = 'v', markersize = 10)
plt.show()
```

```python
# plt.plot(x,y,'선색,마커모양,선의 모양')
plt.plot(x,y,'bo--')
plt.show()
```

```python
# Quiz 다음은 3일 간의 코로나 백신 종류별 접종 인구를 나타낸 그래프이다.
# 그래프를 그리는 코드를 작성하세요
# 단 백신의 종류는 아스트라제네카(az), 화이자(pfizer), 모더나(moderna) 이다.
days = [1,2,3]
az = [2,4,6]
pfizer = [5,1,3]
moderna = [1,2,5]

plt.plot(days, az, label='az')
plt.plot(days, pfizer, label='pfizer', marker ='o', linestyle='--')
plt.plot(days, moderna, label='moderna', marker ='s', ls='-.')
plt.legend()
plt.show()
```

```python
# 막대 그래프
labels = ['A', 'B', 'C']
values = [165,168,170]
plt.ylim(160,180)
plt.bar(labels,values)
plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [165,168,170]
colors = ['r','g','b']
plt.ylim(160,180)
plt.bar(labels,values, color = colors,
        alpha = 0.5, width = 0.5) # alpha: 투명도 지정
plt.show()
```

```python
# 원 그래프
values = [35,20,15,15,10,5]
labels = ['Python','Java','Javascript','C#', 'C/C++', 'ETC']
plt.pie(values,labels=labels,autopct='%.1f%%',
        startangle=90,counterclock=False)
plt.show()
```

```python
values = [35,20,15,15,10,5]
labels = ['Python','Java','Javascript','C#', 'C/C++', 'ETC']
explode = [0.05]*6
plt.pie(values,labels=labels,explode=explode,
        startangle=90,counterclock=False)
plt.legend(loc=(1.2,0.3))
plt.show()
```
