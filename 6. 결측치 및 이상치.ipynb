```python
# 데이터 이상치(Data Outlier)
# 데이터가 정상의 범주에서 벗어난 값
# 변수의 분포 상 비정상적으로 극단적인 값
# 오류 VS 실제적인 값
# 평균과 분산에 영향을 미쳐 통계 결과값의 신뢰성 저하, 왜곡 발생
from seaborn import load_dataset
import pandas as pd
car_df = load_dataset('car_crashes')
car_df.head()
```

```python
# 범위 지정 : 4분위 수 사용
# - IQR 산정
# - 상한값, 하한값 구하기
# - 이상치 중앙값으로 대체
q1 = car_df['alcohol'].quantile(0.25)
q2 = car_df['alcohol'].quantile(0.75)
print(q1,q2)
```

```python
iqr = q2-q1
print(iqr)
```

```python
btom_cut_off = q1 - 1.5*iqr
top_cut_off = q2 + 1.5*iqr
print("하한값: ", btom_cut_off, "상한값: ", top_cut_off)
```

```python
btom_cut_off = "{:.4f}".format(btom_cut_off)
top_cut_off = "{:.4f}".format(top_cut_off)
print("하한값: ", btom_cut_off, "상한값: ", top_cut_off)
```

```python
print(type(btom_cut_off),type(top_cut_off))
```

```python
btom_cut_off = float(btom_cut_off)
top_cut_off = float(top_cut_off)
```

```python
# True = 이상값
(car_df['alcohol']<=btom_cut_off)|(car_df['alcohol']>=top_cut_off)[:5]
```

```python
car_df['alcohol'][outlier_index] = median
car_df.loc[outlier_index,'alcohol'] = median
```
