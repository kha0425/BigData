01.	주어진 trash bag 데이터 세트는 지역별 종량제 봉투 가격을 나타낸다. 가격 컬럼은 각 행의 조건을 만족하는 해당 용량의 종량제 봉투가 존재하면 가격을 값으로, 존재하지 않으면 0을 값으로 갖는다. 이때 용도가 ‘음식물쓰레기’이고 사용 대상이 ‘가정용’인 2L 봉투 가격의 평균을 소수점을 버린 후 정수로 출력하시오.

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
df1 = pd.read_csv('./data/5_1_trash_bag.csv',encoding = 'cp949')
df1.head()
```

```python
# 조건에 맞는 데이터 추출
filtered_data = df1[(df1['용도']=='음식물쓰레기')&(df1['사용대상']=='가정용')]
filtered_data.head()
```

```python
# 2L가격 열에서 가격이 0이 아닌 데이터의 평균 계산 및 정수 출력
average_price = filtered_data[filtered_data['2L가격'] != 0]['2L가격'].mean()
result = int(average_price)
print(result)
```

**2) 작업형 제1유형**

02.	BMI 지수는 몸무게(kg)를 키(m)의 제곱으로 나누어 구하며, BMI값에 따른 비만도 분류는 다음과 같다. 이 때 주어진 bmi 데이터 세트에서 비만도가 정상에 속하는 인원 수와 과체중에 속하는 인원 수의 차이를 정수로 출력하시오.

```python
# 라이브러리 및 데이터 불러오기
df2 = pd.read_csv('./data/5_2_bmi.csv')
df2.head()
```

```python
# BMI 지수 계산
df2['BMI'] = df2['Weight']/((df2['Height']/100)**2)
df2.head()
```

```python
# '정상' 범주와 '과체중' 범주에 해당하는 인원 수 계산
# '정상' 범주 인원 수
normal = df2[(df2['BMI'] >= 18.5)&(df2['BMI'] < 23)].shape[0]
# '과체중' 범주 인원 수
overweight = df2[(df2['BMI'] >= 23)&(df2['BMI']<25)].shape[0]
```

```python
# '정상'과 '과체중' 사이의 차이 계산 및 결과 출력
result = normal - overweight
print(result)
```

**2) 작업형 제1유형**

03.	주어진 students 데이터 세트는 각 학교의 학년별 총 전입학생, 총 전출학생, 전체 학생 수를 나타낸다. 순 전입학생 수는 총 전입학생 수에서 총 전출학생 수를 빼서 구할 수 있다. 순 전입학생이 가장 많은 학교의 전체 학생 수를 구하시오.

```python
# 라이브러리 및 데이터 불러오기
df3 = pd.read_csv('./data/5_3_students.csv', encoding = 'cp949')
df3.head()
```

```python
# 순 전입학생 수 계산
df3['순 전입학생'] = df3['총 전입학생'] - df3['총 전출학생']
df3.head()
```

```python
# 학교별 순 전입학생의 합계와 전체 학생 수의 합계 계산
school_summary = df3.groupby('학교').agg({'순 전입학생':'sum','전체 학생 수':'sum'}).reset_index()
school_summary
```

```python
# 순 전입학생이 가장 많은 학교의 전체 학생 수 출력
result = school_summary.loc[school_summary['순 전입학생'].idxmax(),'전체 학생 수']
print(result)
```

```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import OneHotEncoder
```

```python
# 데이터 불러오기
X_train = pd.read_csv('./data/5_4_x_train.csv')
y_train = pd.read_csv('./data/5_4_y_train.csv')['price']
X_test = pd.read_csv('./data/5_4_x_test.csv')
```

```python
# 원 핫 인코딩
# get_dummies 함수 사용
# drop_first=True 옵션은 각 범주형 변수에서 첫 번째 범주를 제외시켜 더미 변수의 함정(dummy variable trap)을 피할 수 있게 해준다.

X_train_preprocessed = pd.get_dummies(X_train, drop_first= True)
X_test_preprocessed= pd.get_dummies(X_test, drop_first=True)
```

```python
X_train_preprocessed.head()
```

```python
# 훈련 세트에 있는 컬럼과 테스트 세트에 있는 컬럼을 맞추기
X_test_preprocessed = X_test_preprocessed.reindex(columns = X_train_preprocessed, fill_value=0)
```

```python
# 모델 훈련
linear_model = LinearRegression()
random_forest_model = RandomForestRegressor()
nn_model = MLPRegressor(max_iter=500)

linear_model.fit(X_train_preprocessed,y_train)
random_forest_model.fit(X_train_preprocessed, y_train)
nn_model.fit(X_train_preprocessed, y_train)
```

```python
# 모델 평가
y_pred_linear = linear_model.predict(X_train_preprocessed)
y_pred_rf = random_forest_model.predict(X_train_preprocessed)
y_pred_nn = nn_model.predict(X_train_preprocessed)

rmse_linear = np.sqrt(mean_squared_error(y_train,y_pred_linear))
rmse_rf = np.sqrt(mean_squared_error(y_train,y_pred_rf))
rmse_nn = np.sqrt(mean_squared_error(y_train,y_pred_nn))

print(rmse_linear,rmse_rf,rmse_nn)
```

```python
# 예측 결과를 데이터프레임 형식으로 변환
results = pd.DataFrame({'id':range(1,len(y_pred_rf)+1),'price':y_pred_rf})
print(results)
```
