```python
# 인공신경망
# - 인간 두뇌의 신경세포인 뉴런을 모방한 기계학습
```

```python
# 인공신경망 분류
# sklearn.neural_network.MLPClassifier()
# hiddin_layer_sizes : 은닉층 크기(은닉층 수, 뉴런 수)(default=(100,))
# activation : 'relu'(default)
#              'tanh', 'logistic', 'identify'
# solver : 'adam'(default)
#          'sgd', 'lbfgs'
# batch_size : 수렴할 때까지 걸리는 최대 반복 횟수(default = 100)
# learning_rate_init : 학습률 초깃값(default = 0.001)
# max_iter : 에포크 횟수(default = 200)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X_train,X_test,y_train,y_test=train_test_split(iris.data,iris.target,test_size=0.2,random_state=1234,stratify=iris.target)
```

```python
ann_C_1 = MLPClassifier()
model_ann_C_1 = ann_C_1.fit(X_train,y_train)
```

```python
ann_C_2 = MLPClassifier(activation = 'relu')
model_ann_C_2 = ann_C_2.fit(X_train,y_train)
```

```python
ann_C_3 = MLPClassifier(activation = 'tanh', max_iter = 500)
model_ann_C_3 = ann_C_3.fit(X_train,y_train)
```

```python
from sklearn.metrics import f1_score
y_pred_1 = model_ann_C_1.predict(X_test)
y_pred_2 = model_ann_C_2.predict(X_test)
y_pred_3 = model_ann_C_3.predict(X_test)
```

```python
macro_f1_1 = f1_score(y_test,y_pred_1,average = 'macro')
macro_f1_2 = f1_score(y_test,y_pred_2,average = 'macro')
macro_f1_3 = f1_score(y_test,y_pred_3,average = 'macro')
```

```python
print(macro_f1_1,macro_f1_2,macro_f1_3)
```

```python
# 인공신경망 회귀
# sklearn.neural_network.MLPRegressor()
# hiddin_layer_sizes : 은닉층 크기(은닉층 수, 뉴런 수)(default=(100,))
# activation : 'relu'(default)
#              'tanh', 'logistic', 'identify'
# solver : 'adam'(default)
#          'sgd', 'lbfgs'
# batch_size : 수렴할 때까지 걸리는 최대 반복 횟수(default = 100)
# learning_rate_init : 학습률 초깃값(default = 0.001)
# max_iter : 에포크 횟수(default = 200)
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
X_train,X_test,y_train,y_test = train_test_split(diabetes.data,diabetes.target,test_size=0.2,random_state=1234)
```

```python
ann_R_1 = MLPRegressor()
model_ann_R_1 = ann_R_1.fit(X_train,y_train)
```

```python
ann_R_2 = MLPRegressor(activation='relu')
model_ann_R_2 = ann_R_2.fit(X_train,y_train)
```

```python
ann_R_3 = MLPRegressor(activation='tanh', max_iter = 500)
model_ann_R_3 = ann_R_3.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error
y_pred_1 = model_ann_R_1.predict(X_test)
y_pred_2 = model_ann_R_2.predict(X_test)
y_pred_3 = model_ann_R_3.predict(X_test)
```

```python
mae_1 = mean_absolute_error(y_test,y_pred_1)
mae_2 = mean_absolute_error(y_test,y_pred_2)
mae_3 = mean_absolute_error(y_test,y_pred_3)
print(mae_1,mae_2,mae_3)
```
