```python
# 출동시간 평균으로 내림차순 정렬하기
df1_response = df1_response.sort_values('출동시간',ascending = False)
df1_response.head()
```

```python
# 출동시간 평균이 가장 긴 소방서 탐색
results = df1_response.iloc[0,0]
print(results)
```

**2) 작업형 제2유형**

01.	다음은 선호하는 메타버스 유형에 대한 데이터 세트이다. 주어진 훈련 데이터 세트를 활용하여 선호하는 메타버스 유형을 분류하는 모델을 구축하고 테스트 데이터 세트를 대상으로 분류한 결과를 csv 파일로 제출하시오.
* 결과 제출 양식 : 제출한 예측값의 macro_f1 결과를 통해 영역별 배점에 따라 최종 점수가 반영될 예정


```python
# 라이브러리 불러오기
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import f1_score
```

```python
# 라이브러리 불러오기
train_data = pd.read_csv('./data/6_2_1_Metaverse Training Data.csv')
test_data_x = pd.read_csv('./data/6_2_2_Metaverse Testing Data_x.csv')
test_data_y = pd.read_csv('./data/6_2_3_Metaverse Testing Data_y.csv',encoding='cp949')
```

```python
# 데이터 분리
X_train = train_data.drop('Preferred Metaverse Type', axis=1)
y_train = train_data['Preferred Metaverse Type']
X_test = test_data_x.drop('id', axis=1)
y_test = test_data_y['Preferred Metaverse Type']
```

```python
# 변수 확인
X_train.head()
```

```python
# 수치형 데이터 스케일링
scaler = StandardScaler()
numeric_features = X_train.select_dtypes(include=['number']).columns
X_train[numeric_features] = scaler.fit_transform(X_train[numeric_features])
X_test[numeric_features] = scaler.transform(X_test[numeric_features])
```

```python
# 범주형 데이터 인코딩
encoder = OneHotEncoder()
categorical_features = X_train.select_dtypes(include=['object']).columns
X_train_encoded = encoder.fit_transform(X_train[categorical_features])
X_test_encoded = encoder.transform(X_test[categorical_features])
print(X_train_encoded.shape,X_test_encoded.shape)
```

```python
# 인코딩된 범주형 데이터를 DataFrame으로 변환
X_train_encoded_df = pd.DataFrame(X_train_encoded.toarray(), columns=encoder.get_feature_names_out(categorical_features))
X_test_encoded_df = pd.DataFrame(X_test_encoded.toarray(), columns=encoder.get_feature_names_out(categorical_features))

```

```python
# 수치형 데이터와 범주형 데이터를 결합
X_train_final = pd.concat([X_train[numeric_features].reset_index(drop=True), X_train_encoded_df.reset_index(drop=True)], axis=1)
X_test_final = pd.concat([X_test[numeric_features].reset_index(drop=True), X_test_encoded_df.reset_index(drop=True)], axis=1)
```

```python
X_train_final.head()
```

```python
# 로지스틱 회귀 모델 생성 및 평가
# 로지스틱 회귀 모델 초기화 및 훈련
lr_model = LogisticRegression()
lr_model.fit(X_train_final, y_train)
# 테스트 데이터에 대한 예측
lr_predictions = lr_model.predict(X_test_final)

# Macro F1 Score 계산
lr_macro_f1 = f1_score(test_data_y['Preferred Metaverse Type'], lr_predictions, average='macro')
print(f"로지스틱 회귀 모델 Macro F1 Score: {lr_macro_f1}")
```

```python
# 랜덤 포레스트 모델 생성 및 평가
# 랜덤 포레스트 모델 초기화 및 훈련
rf_model = RandomForestClassifier()
rf_model.fit(X_train_final, y_train)

# 테스트 데이터에 대한 예측
rf_predictions = rf_model.predict(X_test_final)

# Macro F1 Score 계산
rf_macro_f1 = f1_score(test_data_y['Preferred Metaverse Type'], rf_predictions, average='macro')
print(f"랜덤 포레스트 모델 Macro F1 Score: {rf_macro_f1}")
```

```python
# SVM 모델 생성 및 평가
# SVM 모델 초기화 및 훈련
svm_model = SVC()
svm_model.fit(X_train_final, y_train)

# 테스트 데이터에 대한 예측
svm_predictions = svm_model.predict(X_test_final)

# Macro F1 Score 계산
svm_macro_f1 = f1_score(test_data_y['Preferred Metaverse Type'], svm_predictions, average='macro')
print(f"SVM 모델 Macro F1 Score: {svm_macro_f1}")
```

```python
# 최종 모델 선택 및 파일 저장
# 최종 예측
final_predictions = svm_model.predict(X_test_final)

# CSV 파일로 저장
results = pd.DataFrame(final_predictions, columns=['Predicted'])
results.to_csv('./output/000001.csv', index=False)
```

**3) 작업형 제3유형**

01.	A 도시의 교통수단은 버스, 지하철, 택시, 자전거가 있다. 전문가들은 각 교통수단의 실제 선택 비율이 각각 40%, 30%, 20%, 10%라고 예측했다. 이를 검증하기 위해 대중교통 이용객들이 선호하는 교통수단에 대해 설문조사를 했다. 이를 위해 수립한 귀무가설과 대립가설은 아래와 같다. 이에 대한 검정을 진행하시오.



귀무가설 (H0) 	대중교통 이용객의 실제 교통수단 선택 비율이 예상 비율과 같다.


대립가설 (H1) 	대중교통 이용객의 실제 교통수단 선택 비율이 예상 비율과 다르다.



(1)	주어진 설문조사 데이터에서 전체 건수 대비 Subway를 선호하는 이용자의 비율을 구하시오. (단, 반올림하여 소수점 둘째 자리까지 계산한다.)

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
df5 = pd.read_csv('./data/6_3_1_transport_data.csv', encoding='cp949')
df5.head()
```

```python
# Subway를 선호하는 이용자의 비율 계산 (반올림하여 소수 둘째 자리까지 계산)
# 지하철을 선호하는 사용자 비율 계산
subway_pref_rate = (df5['TransportMode'] == 'subway').mean()

# 소수점 두 자리까지 반올림
results = round(subway_pref_rate, 2)
print(results)
```

(2)	예상 이용자 수와 실제 이용자 수가 동일한 비율을 보이는지 적합도를 검정하고 검정 통계량을 소수 둘째 자리로 반올림하시오.


```python
# 예산 이용자 수와 실제 이용자 수의 비율 계산

# 라이브러리 불러오기
from scipy.stats import chisquare

# 각 교통 수단별 사용자 수
observed_counts = df5['TransportMode'].value_counts()

# 관찰된 순서를 기대 순서와 일치시키기 위해 정렬
observed_counts = observed_counts.reindex(['bus', 'subway', 'taxi', 'bicycle']).fillna(0)

# 전체 응답자 수
total_responses = df5.shape[0]

# 기대 비율에 따른 기대 사용자 수
expected_rates = [0.4, 0.3, 0.2, 0.1]  # 각 교통 수단별 기대 비율

#기대 빈도 구하기
expected_counts = [rate * total_responses for rate in expected_rates]
```

```python
# 적합도 검정 수행
chi2_stat, p_value = chisquare(f_obs=observed_counts, f_exp=expected_counts)

# 검정 통계량 반올림
results = round(chi2_stat, 2)

# 결과 출력
print(results)
```

 (3)	위의 통계량에 대한 p-value를 구하고, 유의수준 0.05하에서 가설 검정의 결과를 채택, 기각 중 하나로 선택하시오 (단, p-value 값은 반올림하여 소수점 둘째 자리까지 계산한다.)


```python
# p-값 반올림
p_value_rounded = round(p_value, 2)

# 유의 수준 0.05와 비교하여 결과 결정
if p_value_rounded > 0.05:
    decision = "귀무 가설 수용: 실제 교통 수단 선택 비율이 기대 비율과 유의미하게 다르지 않음"
else:
    decision = "귀무 가설 기각: 실제 교통 수단 선택 비율이 기대 비율과 유의미하게 다름"

# 결과 출력
print(f"p-값: {p_value_rounded}, 결정: {decision}")
```

3.  작업형 제3유형
 02.	물고기의 무게를 다중회귀분석으로 분석하고자 한다. 데이터의 구성은 아래 표와 같다. Weight를 Y변수로 Length, Height, Width, FinSize를 X변수로 하는 다중회귀분석을 수행하고 아래에 답하시오.  


(1)	Length의 회귀계수를 구하고, 소수 둘째 자리로 반올림하시오.

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
import statsmodels.api as sm


df6 = pd.read_csv('./data/6_3_2_fish_weight_data.csv')
df6.head()
```

```python
# 독립변수 및 종속변수 설정

X = df6[['Length', 'Height', 'Width', 'FinSize']]  # 독립 변수
Y = df6['Weight']  # 종속 변수

# 상수항을 독립 변수에 추가
X = sm.add_constant(X)
```

```python
# 다중 회귀 모델 생성 및 적합
model = sm.OLS(Y, X).fit()
print(model.params)
```

```python
# Length의 회귀 계수를 구하고 소수점 둘째 자리까지 반올림
results = round(model.params['Length'], 2)
print(results)
```

(2)	Height의 회귀계수의 p-value를 구하시오.

```python
# p-value 계산
results = model.pvalues['Height']
print(results)
```

(3)	Width의 회귀계수가 p-value 0.05를 기준으로 했을 때 통계적으로 의미가 있는지를 평가하시오. 의미가 있는 경우엔 ‘Y’, 없는 경우엔 ‘N’을 출력하시오.

```python
results = 'Y' if model.pvalues['Width'] < 0.05 else 'N'
print(results)
```
