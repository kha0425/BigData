01. 다음은 Boston Housing 데이터 세트이다. 범죄율 컬럼인 Crim 항목의 상위에서 10번째 값(즉, 범죄율을 큰 순서로 오름차순 정렬했을 때 10번에 위치한 값)으로 상위 10개의 값을 변환한 후, age가 80 이상인 데이터를 추출하여 crim의 평균값을 계산하시오.

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
boston_df = pd.read_csv('./data/1_1_boston.csv')
boston_df.head()
```

```python
# 내림차순 정렬
boston_df_sorted = boston_df.sort_values('CRIM',ascending = False)
boston_df_sorted.head(10)
```

```python
# 상위 10번째 값 확인 및 변수 할당
df_crim_10 = boston_df_sorted.iloc[9,0]
print(df_crim_10)
```

```python
# 상위 10개의 값을 crim_10으로 변환
boston_df_sorted.iloc[:10,0] = df_crim_10
boston_df_sorted.head(11)
```

```python
# AGE 변수 값이 80 이상인 데이터 추출 후
boston_df_sorted_age_80 = boston_df_sorted[boston_df_sorted['AGE']>=80]
boston_df_sorted_age_80.tail()
boston_df_sorted_age_80['AGE'].min()
```

```python
# CRIM 평균 계산
result = boston_df_sorted_age_80['CRIM'].mean()
print(result)
```

02. 다음은 California Housing 데이터 세트이다. 주어진 데이터의 첫 번째 행부터 순서대로 80%까지의 데이터를 훈련 데이터로 추출 후, 전체 방 개수 컬럼을 의미하는 'total_bedrooms' 변수의 결측치를 'total_bedrooms'변수의 중앙값으로 대체한 데이터 세트를 구성한다. 결측치 대체 전의 'total_bedrooms'변수 표준편차 값과 결측치 대체 후의 'total_bedrooms'변수 표준편차 값의 차이에 대한 절댓값을 계산하시오.

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
df2 = pd.read_csv('./data/1_2_housing.csv')
df2.head()
```

```python
# 전체 데이터 확인
df2.describe()
df2.info()
```

```python
# total_bedrooms 컬럼 80% 추출
df2_80 = df2['total_bedrooms'][:int(len(df2)*0.8)] # 소수점이 나올수 있기 때문에 int 함수 사용
print(len(df2))
print(len(df2_80))
```

```python
# 80% 추출 후 결측치 확인
df2_80.isnull().sum()
```

```python
# 결측치 대체 전 표준편차
std_before = df2_80.std() # quantile(0.25),var,median,mean,std
print(std_before)
```

```python
# 중앙값
median = df2_80.median()
print(median)
```

```python
# 중앙값으로 결측치 대체
df2_80_fill = df2_80.fillna(median)
```

```python
# 결측치 대체 후 표준편차
std_after = df2_80_fill.std()
print(std_after)
```

```python
# 표준편차 값 차이 (절대값)
result = abs(std_before-std_after)
print(result)
```

3. 2번 문항에서 활용한 California Housing 데이터 세트를 그대로 활용한다. 인구 컬럼인 population 항목의 이상값의 합계를 계산하시오. (이상값은 평균에서 1.5 * 표준편차를 초과하거나 미만인 값의 범위로 정한다.

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
df3 = pd.read_csv('./data/1_3_housing.csv')
df3.head()
```

```python
# population 컬럼 분리 후 평균 및 표준편차 구하기
df3_popu = df3['population']
mean = df3_popu.mean()
std = df3_popu.std()
print(mean)
print(std)
```

```python
# 이상값 범위 : 상한값, 하한값
top = mean + (std*1.5)
btom = mean - (std*1.5)
print(top)
print(btom)
```

```python
# 이상값 데이터 추출
population_outlier = df3_popu[(df3_popu>top)|(df3_popu<btom)]
print(population_outlier)
```

```python
result = population_outlier.sum()
print(result)
```

1. 다음은 E-commerce Shipping 데이터 세트이다. 주어진 훈련 데이터를 활용하여 고객이 주문한 물품의 정시 도착 여부를 예측하고 해당 예측 결과를 csv 파일로 제출하시오.

- 결과 제출 양식 : 제출한 예측값의 ROC_AUC 점수 결과를 통해 영역별 배점에 따라 최종 점수가 반영될 예정

- 변수 설명

> ID : 고객 ID 번호

> Warehouse_block : 창고의 블록 단위 구역(A, B, C, D, E, F)

> Mode_of_shipment : 제품 배송 방법

> Customer_care_calls : 문의 전화 수

> Customer_rating : 고객의 등급 (1: 가장 낮음, 5: 가장 높음)

> Cost_of_the_Product : 제품의 비용(달러 기준)

> Prior_purchases : 사전 구매 수량

> Product_importance : 제품의 중요도

> Gender : 성별 (F :여성, M : 남성)

> Discount_offered : 할인혜택

> Weight_in_gas : 그램 단위 무게

> Reached.on.Time_Y.N : 정시도착 유무

[유의사항]
- 성능이 우수한 예측 모형을 구축하기 위해서는 적절한 데이터 전처리, 피처 엔지니어링, 분류 알고리즘, 하이퍼파라미터 튜닝, 모형 앙상블 등이 수반되어야 한다.
- 수험번호.csv (예 : 0000.csv) 파일이 만들어지도록 코드를 제출한다.
- 제출한 모델의 성능은 ROC-AUC로 평가지표에 따라 채점한다.

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
X_train = pd.read_csv('./data/1_1_commerce_x_train.csv')
X_test = pd.read_csv('./data/1_2_commerce_x_test.csv')
y_train = pd.read_csv('./data/1_3_commerce_y_train.csv')
```

```python
# object 컬럼 삭제
X_train = X_train.drop(['Warehouse_block','Mode_of_Shipment','Product_importance','Gender'],axis=1)
X_test = X_test.drop(['Warehouse_block','Mode_of_Shipment','Product_importance','Gender'],axis=1)
```

```python
X_train.info()
```

```python
# 모델링 및 평가 라이브러리 임포트
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
```

```python
# 모델링을 위한 ID 변수 제거
X_train_id = X_train.pop('ID')
X_test_id = X_test.pop('ID')
```

```python
X_train.head()
```

```python
X_train_id
print(type(X_train_id),X_train_id.ndim,X_train_id.shape)
```

```python
# 훈련, 검증 데이터 분할
from sklearn.model_selection import train_test_split
X_tr,X_val,y_tr,y_val = train_test_split(X_train,y_train['Reached.on.Time_Y.N'],test_size = 0.2,random_state = 1234)
```

```python
# 로지스틱회귀분석
model_lgm = LogisticRegression()
model_lgm.fit(X_tr,y_tr)
pred_lgm = model_lgm.predict_proba(X_val)
roc_auc_score(y_val,pred_lgm[:,1])
```

```python
# 의사결정나무
model_dt = DecisionTreeClassifier()
model_dt.fit(X_tr,y_tr)
pred_dt = model_dt.predict_proba(X_val)
roc_auc_score(y_val,pred_dt[:,1])
```

```python
# 랜덤포레스트
model_rf = RandomForestClassifier()
model_rf.fit(X_tr,y_tr)
pred_rf = model_rf.predict_proba(X_val)
roc_auc_score(y_val,pred_rf[:,1])
```

```python
# 최종 모델 선택
model_final = RandomForestClassifier()
model_final.fit(X_train,y_train['Reached.on.Time_Y.N'])
pred_final = model_final.predict_proba(X_test)
pred_final
```

```python
# 제출 파일 생성
submission = pd.DataFrame({'ID' : X_test_id,'Reached.on.Time_Y.N':pred_final[:,1]})
submission.head()
```

```python
# 저장 : 수험번호.csv
submission.to_csv('./output/submission.csv',index=False)
```
