```python
# Quix 01
# 정수를 입력 받은 후 출력하는 프로그램을 설계하려고 한다.
# 단, 숫자 0을 입력 받으면 프로그램이 종료하는 프로그램 코드를 작성하여라
# (while, break 문 사)
a = int(input("정수를 입력하세요.(0을 누르면 종료)"))
while True:
  print("입력한 숫자는:",a)
  if a == 0:
    break
  a = int(input("정수를 입력하세요.(0을 누르면 종료)"))
print("종료")
```

```python
# Quiz 2
# 정수를 입력 받은 후 출력하는 프로그램을 설계하려고 한다.
# 단, 숫자 0을 입력받으면 프로그램이 종료하고
# 음수를 입력받으면 "음수입니다."를 출력하고
# 다시 프로그램이 실행되는 코드를 작성하여라
number = int(input("정수를 입력하세요(0을 누르면 종료):"))
while True:
  if number > 0:
    print("입력한 숫자는:",number)
    number = int(input("정수를 입력하세요(0을 누르면 종료):"))
  if number == 0:
    break
  if number < 0:
    print("입력한 숫자는:",number)
    print("음수 입니다.")
    number = int(input("정수를 입력하세요(0을 누르면 종료):"))
    continue
print("종료")
```
