```python
# 표준화(Standaridzation)
# - 각 데이터가 평균을 기준으로 얼마나 떨어져 있는지를 나타내는 값으로 변환
# - 평균이 0, 표준편차가 1인 분포로 변환
# - 표준화와 정규화는 데이터 전처리에서 상호 교환하여 사용
# - scikit-learn preprocessing 모듈의 StandardScaler 클래스 사용
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()
iris
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
iris.data
```

```python
iris_data = iris.data
iris_df = pd.DataFrame(data = iris_data, columns = iris.feature_names) # 행과 열 정의
```

```python
iris_df.head()
```

```python
print("평균: \n", iris_df.mean())
print("분산: \n", iris_df.var())
```

```python
# 표준화
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() # 객체 생성
```

```python
scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df) # 표준화 변환
```

```python
# 표준화 : 평균이 0, 표준편차 1
iris_scaled
```

```python
iris_df_scaled = pd.DataFrame(data = iris_scaled, columns = iris.feature_names)
```

```python
iris_df_scaled.head()
```

```python
print("평균: \n", iris_df_scaled.mean())
print("평균: \n", iris_df_scaled.var())
```

```python
# 정규화(Normarlization)
# - 데이터의 범위를 0 ~ 1로 변환
# - scikit-learn preprocessing 모듈의 MinMaxScaler 사용
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
print("최소값: \n", iris_df.min())
print("최대값: \n", iris_df.max())
```

```python
scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)
```

```python
iris_df_scaled = pd.DataFrame(data=iris_scaled,columns=iris.feature_names)
```

```python
print("최소값: \n", iris_df_scaled.min())
print("최대값: \n", iris_df_scaled.max())
```
